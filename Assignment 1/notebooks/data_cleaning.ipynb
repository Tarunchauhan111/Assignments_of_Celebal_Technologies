**2. Handle the Missing Values in the Dataset**

Before Handling the data we need a copy of the data

```python
df_copy = df.copy()
```

```python
df_copy
```

Identifying the Missing Values

```python
df_copy.isnull().sum()
```

Filling the Missing Values

```python
df_copy["seller_name"].fillna("Unknown Seller")
```

```python
df_copy["videos"].fillna("No Video")
```

```python
df_copy["what_customers_said"].fillna("No Review")
```

```python
df_copy["seller_information"].fillna("Not Available")
```

**3. Perform Basic Operations**

Select Important Columns

```python
selected_df = df_copy[["product_id","title","rating","final_price","category"]]
selected_df.head()
```

Filter Highly Rated Products

```python
high_rating = df_copy[df_copy["rating"] > 4.5]
high_rating[["title", "rating"]]
```

Filter Products by Category

```python
backpacks = df_copy[df_copy["category"] == "backpacks"]
backpacks.head()
```

**4. Remove Duplicates**

Check Duplicate Rows

```python
(df_copy.duplicated().sum())
```

There is no duplicated rows in the data But lets run the command for Example

```python
df_copy.drop_duplicates(inplace=True)
```

**5. Clean Price Column**

Remove Currency Symbols

```python
df_copy["final_price"] = (df_copy["final_price"].replace(r'[₹,"]', '', regex=True).astype(float))
```

**6. Creating the Derived Column**

Create a Quantity Column

```python
df_copy["quantity"] = 1
```

Create Total Amount

```python
df_copy["total_amount"] = df_copy["final_price"] * df_copy["quantity"]
```

Let's check the output

```python
df_copy[["title", "final_price", "quantity", "total_amount"]].head()
```
