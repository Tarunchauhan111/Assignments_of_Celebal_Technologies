**Importing Spark Session and functions**

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import DecimalType
```

**Starting the Spark Session**

```python
spark = SparkSession.builder.appName("Spark Basics Assignment").getOrCreate()
print("Spark Started Successfully")
```

**Read the CSV file**

```python
df = spark.read.csv( "C:/Users/hp/OneDrive/Desktop/Assignment/Assignment 6/data6/dataset.csv", header=True, inferSchema=True )
print("Dataset Loaded Successfully")
df.show(5)
```

Schema of the Table

```python
df.printSchema()
```

Printing the essential Columns

```python
df = df.select( "Student_ID", "Major_Category", "Pre_Semester_GPA", "Weekly_GenAI_Hours", "Traditional_Study_Hours", "Post_Semester_GPA", "Skill_Retention_Score", "Anxiety_Level_During_Exams", "Burnout_Risk_Level" )
df.show()
```

**Filtering**

1. Filtering the data whose Traditional_Study_Hours > 15

```python
df.filter(df.Traditional_Study_Hours > 15).show()
```

2. Filtering the data whose Weekly_GenAI_Hours > 10

```python
df.filter(df["Weekly_GenAI_Hours"] > 10).show()
```

3. Filtering the data whose Post_Semester_GPA > 3

```python
df.filter(df["Post_Semester_GPA"] > 3).show()
```

**Grouping**

1. Finding the total_students for Major_Categories

```python
df.groupBy("Major_Category").agg(count("*").alias("Total_Students")).show()
```

2. Find the average pre_semester_GPA for each major_categories

```python
df.groupBy("Major_Category").agg(avg("Pre_Semester_GPA").alias("Average_Pre_GPA")).show()
```

**Renaming the Columns**

```python
df = df.withColumnRenamed( "Weekly_GenAI_Hours", "GenAI_Hours" )
df = df.withColumnRenamed( "Traditional_Study_Hours", "Study_Hours" )
df = df.withColumnRenamed( "Pre_Semester_GPA", "Pre_GPA" )
df = df.withColumnRenamed( "Post_Semester_GPA", "Post_GPA" ) 
df.printSchema()
```

**Changing the data type of few columns**

```python
df = df.withColumn( "GenAI_Hours", col("GenAI_Hours").cast("int"))
df = df.withColumn( "Pre_GPA", col("Pre_GPA").cast(DecimalType(3,2)))
df = df.withColumn( "Post_GPA", col("Post_GPA").cast(DecimalType(3,2)))
```

```python
df = df.withColumn( "GPA_Improvement", round( col("Post_GPA") - col("Pre_GPA"), 2))
df = df.withColumn( "Anxiety_Level_During_Exams", col("Anxiety_Level_During_Exams").cast("double"))
df.printSchema()
```

**Add new columns**

```python
df = df.withColumn( "AI_Usage", when(col("GenAI_Hours") > 12,"Heavy").when(col("GenAI_Hours") > 6,"Moderate").otherwise("Light"))
df = df.withColumn( "Study_AI_Ratio", round( try_divide( col("Study_Hours"), col("GenAI_Hours")), 2))
df.show()
```

3. find the average post_gpa for every major category

```python
df.groupBy("Major_Category").avg("Post_GPA").show()
```

4. find the average genAI hours for every major category

```python
df.groupBy("Major_Category").avg("GenAI_Hours").show()
```

5. find the average skill retention score for every major category

```python
df.groupBy("Major_Category").avg("Skill_Retention_Score").show()
```

**Total no. of rows**

```python
df.count()
```

**Checking the table after some transformations**

```python
df.show()
```

**Saving the CSV file in Parquet**

```python
df.write.mode("overwrite").parquet("C:\\Users\\hp\\OneDrive\\Desktop\\Assignment\\Assignment 6\\output6\\output_parquet")
```

**Reading the Parquet file**

```python
parquet_df = spark.read.parquet("C:\\Users\\hp\\OneDrive\\Desktop\\Assignment\\Assignment 6\\output6\\output_parquet")
parquet_df.show()
```

**Checking the null values**

```python
parquet_df.select([ sum(col(c).isNull().cast("int")).alias(c) for c in parquet_df.columns ]).describe().show()
```

**Removing the null values**

```python
parquet_df = parquet_df.dropna()
```

**Checking the number of rows after removing the null values**

```python
print("Total Rows:", parquet_df.count())
```

**Filtering the data whose GenAI_Hours > 5**

```python
parquet_df = parquet_df.filter( col("GenAI_Hours") > 5).show()
```

**Show the data in descending order by GPA_Improvement**

```python
parquet_df = parquet_df.orderBy( col("GPA_Improvement").desc())
```

**Save the CSV file**

```python
parquet_df.write.mode("overwrite").option("header", True).csv("C:\\Users\\hp\\OneDrive\\Desktop\\Assignment\\Assignment 6\\output6\\csv_output")
```
